```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(loc = 20, scale = 5, size = 150)

df = pd.Series(data)
df
```

```python
xdata = np.linspace(0, 149, 150)
len(xdata)
```

```python
xdata
```

```python
plt.plot(data)
```

```python
plt.plot(df.ewm(alpha=0.05, adjust=False).mean())
plt.savefig('Log_avg_alfa_0.1.jpg', dpi=300)  
```

```python
plt.scatter(xdata, df.ewm(alpha=0.05, adjust=False).mean())
plt.savefig('Log_avg_scatter_alfa_0.3.jpg', dpi=150, pad_inches=0.1)  
```

```python
x1 = np.linspace(0.0, 5.0, 100)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.sin(x1)
fig, ax = plt.subplots(figsize=(6, 4))
fig.subplots_adjust(bottom=0.15, left=0.2)
ax.plot(x1, y1)
ax.plot(x1,y2)
ax.set_xlabel('time [s]')
ax.set_ylabel('Damped oscillation [V]')
ax.set_title('Title with loc at')
plt.savefig('Image with annotations.jpg', dpi=150, pad_inches=0.1)  
plt.show()
```
